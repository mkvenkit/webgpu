```python
# Reimport necessary libraries and redefine functions after code reset
import matplotlib.pyplot as plt
import math

# Redefine necessary functions and variables
def convolve_1d_with_padding(signal, kernel):
    kernel_size = len(kernel)
    signal_size = len(signal)
    padding = kernel_size // 2
    padded_signal = [0] * padding + signal + [0] * padding
    result = []

    # Apply convolution with zero padding
    for i in range(signal_size):
        weighted_sum = 0
        for j in range(-padding, padding + 1):
            weighted_sum += padded_signal[i + j + padding] * kernel[j + padding]
        result.append(weighted_sum)
    
    return result

# Define the Gaussian kernel function
def gaussian_kernel(radius, sigma):
    kernel = [math.exp(-x * x / (2 * sigma * sigma)) for x in range(-radius, radius + 1)]
    normalization_factor = sum(kernel)  # Normalize the kernel
    kernel = [value / normalization_factor for value in kernel]
    return kernel

# Parameters for Gaussian kernel
R = 2  # Radius
s = 1  # Sigma

# Generate the Gaussian kernel
gaussian_kernel_values = gaussian_kernel(R, s)

# Define a 32-pixel step function
step_function_32 = [0] * 16 + [1] * 16

# Apply convolution
gaussian_blurred_step_32 = convolve_1d_with_padding(step_function_32, gaussian_kernel_values)

# Print out the Gaussian kernel values
print("Gaussian kernel values:", gaussian_kernel_values)

# Plot the 32-pixel array (step function) and the convolution result
plt.figure(figsize=(10, 6))

# Plot the original step function as points
plt.subplot(2, 1, 1)
plt.plot(step_function_32, 'o', label="Original 32-Pixel Step Function")
plt.title("Step Function")
plt.xlabel("Pixel index")
plt.ylabel("Value")
plt.grid(True)

# Plot the Gaussian convolution result as points
plt.subplot(2, 1, 2)
plt.plot(gaussian_blurred_step_32, 'o', label="Convolution Result with Gaussian Kernel", color='orange')
plt.title("Convolution Result (Gaussian Kernel)")
plt.xlabel("Pixel index")
plt.ylabel("Value")
plt.grid(True)

plt.tight_layout()
plt.show()
```
